```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from scipy import stats

from google.colab import drive
drive.mount('/content/drive')
```

```python
# Gerar dados de vendas aleatórios (exemplo fictício)

np.random.seed(42)  # Para reprodução dos resultados

num_samples = 100  # Número total de vendas

sales = np.random.normal(loc=50, scale=10, size=num_samples)
sales = sales.round()
# Dados de vendas (distribuição normal)
```

```python
# Gerar dados de dias da semana (exemplo fictício)

days_of_week = ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta',
'Sábado', 'Domingo']

day_labels = np.random.choice(days_of_week, size=num_samples)
```

```python
# Gerando 100 datas aleatórias entre 01/01/2023 e 31/12/2023
start_date = "2023-01-01"
end_date = "2023-12-31"

# Gerando datas aleatorias para as vendas
time_delta = np.random.randint(0, (pd.to_datetime(end_date) - pd.to_datetime(start_date)).days, size=100)
dates = pd.to_datetime(start_date) + pd.to_timedelta(time_delta, unit='D')

# Gerando 100 valores de vendas aleatórios entre 10 e 1000
sales_values = np.random.randint(10, 1001, size=100)
```

```python
# Criar DataFrame com os dados

df = pd.DataFrame({"Vendas": sales, "Dia": day_labels, "Data": dates, "Valor da Venda em R$": sales_values})
```

```python
# Visualizando as 5 primeiras linhas, as ultimas 5 e o tamanho de colunas e linhas
display(df)
```

## Breve descripção dos dados.

### Coluna Vendas: Representa o valor das vendas totais nesse dia em especifico

### Coluna Dia: Representa o dia que foram efetivadas as vendas

### Coluna Data: Representa a Data que foram efetivadas as Vendas

### Valor da Venda: Representa o valor total das Vendas em R$

## Hipotese (0) ou Nula: Não há diferença significativa entre as vendas nos finais de semana e nos dias da semana.

## Hipotese (1): Há diferença significativa entre as vendas nos finais de semana e nos dias da semana

```python
# Separar vendas por finais de semana e dias da semana

weekend_sales = df[df["Dia"].isin(["Sábado", "Domingo"])][["Vendas", "Data", "Valor da Venda em R$"]]

weekday_sales = df[df["Dia"].isin(["Segunda", "Terça", "Quarta",
"Quinta", "Sexta"])][["Vendas", "Data", "Valor da Venda em R$"]]
```

```python
# Realizando Teste Estatistico "T"
t_statistic, p_value = stats.ttest_ind(weekend_sales["Vendas"], weekday_sales["Vendas"])

print(f"t-statistic: {t_statistic}")
print(f"p-value: {p_value}")
```

## Conclusão do Teste: Com base nos resultados do teste T, podemos concluir que o o valor-p é maior que 0.05, o que quer dizer que não há diferença estatísticamente significativa entre as médias de vendas de Eletrõnicos nos finais de semana e nos dias da semana. Ou seja, ainda não podemos rejeitar a Hipoteset (0), teremos que continuar com a Analise
